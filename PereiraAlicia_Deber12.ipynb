### a. 
$$y(t) = \frac{t}{1 + \ln t}$$  


```python
import matplotlib.pyplot as plt
import numpy as np

def f(t, y):
    return y/t-(y/t)**2

a = 1
b = 2
N = 10

y0 = 1

resultado = ODE_euler(a=a, b=b,f=f, y_t0=y0 ,N=N)
resultado

ys = resultado[0]
ts = resultado[1]

t_real = np.linspace(a, b, 100)
y_real = (t_real)/ (1 + np.log(t_real))

plt.figure(figsize=(8, 5))
plt.plot(ts, ys, marker='o', linestyle='-', label='Aproximación Euler', color='violet')
plt.plot(t_real, y_real, linestyle='--', label='Solución Real', color='blue')
plt.title('Gráfica de y vs t usando el Método de Euler', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=10)
plt.show()
```

## Tercer Ejercicio

5. **Utilice los resultados del ejercicio 3 y la interpolación lineal para aproximar los siguientes valores de $y(t)$. Compare las aproximaciones asignadas para los valores reales obtenidos mediante las funciones determinadas en el ejercicio 4.**

### a.
$y(0.25)$ y $y(0.93)$  


```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
    return y/t-(y/t)**2

a = 1
b = 2
N = 10

y0 = 1

resultado = ODE_euler(a=a, b=b,f=f, y_t0=y0 ,N=N)
resultado

y_euler = np.array(resultado[0])
t_euler = np.array(resultado[1])

coeficientes = np.polyfit(t_euler, y_euler, 1)

def interpolar(t, coeficientes):
    return np.polyval(coeficientes, t)

t_grafica = np.linspace(min(t_euler), max(t_euler), 100)
y_grafica = interpolar(t_grafica, coeficientes)

def y_real(t):
    return (t)/ (1 + np.log(t))

t_valores = np.array([0.25, 0.93])
y_reales = y_real(t_valores)
y_interpolados = interpolar(t_valores, coeficientes)

print("Coeficientes de la interpolación:", coeficientes)

errores = abs(y_interpolados - y_reales)
for t_val, y_real_val, y_interp, error in zip(t_valores, y_reales, y_interpolados, errores):
    print(f"Error en y({t_val}): Interpolado = {y_interp}, Real = {y_real_val}, Error = {error}")

plt.scatter(t_euler, y_euler, color='olive', label='Datos reales')
plt.plot(t_grafica, y_grafica, color='skyblue', label='Interpolación lineal')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title('Interpolación Lineal y ODE')
plt.grid(True)
plt.show()
```
