### Sexto Ejercicio

6. **El sistema lineal**

$$
x_1 - x_3 = 0.2, \\
-\frac{1}{2}x_1 + x_2 - \frac{1}{4}x_3 = -1.425, \\
x_1 - \frac{1}{2}x_2 + x_3 = 2.
$$

**tiene la soluci√≥n $(0.9, -0.8, 0.7)$**.


### a. 
¬øLa matriz de coeficientes 

$$
A = 
\begin{bmatrix}
1 & 0 & -1 \\
-\frac{1}{2} & 1 & -\frac{1}{4} \\
1 & -\frac{1}{2} & 1 
\end{bmatrix}
$$

tiene diagonal estrictamente dominante?


```python
import numpy as np

def es_estrictamente_diagonal(matriz):
    filas, columnas = matriz.shape
    if filas != columnas:
        return False  

    for i in range(filas):
        for j in range(columnas):
            if i != j and matriz[i, j] != 0:
                return False  

    return True

A = np.array([[1, 0, -1],[-0.5, 1, -0.25],[1, -0.5, 1]])


print("Matriz A es estrictamente diagonal:", es_estrictamente_diagonal(A)) 
```
